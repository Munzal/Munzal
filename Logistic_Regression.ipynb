```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
plt.scatter(df['age'],df['bought_insurance'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtr, xtt, ytr, ytt = train_test_split(df[['age']], df.bought_insurance, train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear')
```

```python
model.fit(xtr, ytr)
```

```python
model.score(xtr, ytr)
```

```python
y_p = model.predict(xtt)
y_p
```

```python
model.coef_
```

```python
model.intercept_
```

```python
import math
def sigmoid(x):
    return 1 / (1+math.exp(-x))
```

```python
def prediction_function(age):
    z = 0.0548 * age - 1.615
    y = sigmoid(z)
    return y
```

```python
age = 25
prediction_function(age)
```

```python
age = 43
prediction_function(age)
```

```python
for i in range(25,45):
    h = prediction_function(i)
    if(h>0.5):
        print(i)
        break
```
